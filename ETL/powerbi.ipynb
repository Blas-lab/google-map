En este notebook prepararemos los datos para ingestar en PowerBi y realizar luego el análisis por ciudad para efectuar las recomendaciones de instalación de nuevas unidades de negocios a nuestro cliente

```python
# Importamos las librerias necesarias
import pandas as pd
import numpy as np
```

```python
# Importamos el dataset a utilizar con los datos de los comercios
metadata = pd.read_csv(r"datasets_limpios\metadata_final.csv", delimiter=";", names= ["name","address", "gmap_id", "description", "latitude", "longitude", "category", "avg_rating","num_of_reviews","price","hours","misc","relative_results","url"])
```

```python
# Visualizamos el dataframe
metadata
```

```python
# Generamos una nueva columna con el Estado y codigo postal de cada negocio,
# extrayendola del campo de la dirección
metadata['CPST'] = metadata['address'].str.split(',').str[-1]
```

```python
# Generamos una nueva columna con la ciudad, extrayendola del campo dirección
metadata['city'] = metadata['address'].str.split(',').str[-2]
```

```python
# Generamos una columna solo con el CP
metadata["cp"] = metadata["CPST"].str.split(" ").str[-1]
```

```python
# Generamos una columna colo con el Estado
metadata["state"] = metadata["CPST"].str.split(" ").str[-2]
```

```python
# Eliminamos la columna temporal
metadata = metadata.drop(["CPST"],axis=1)
```

```python
# Importamos la tabla de categorìas y grupos
categorias = pd.read_csv(r"datasets_limpios\categorias.csv")
```

```python
# Visualizamos
categorias
```

```python
# Agregamos estos datos al dataframe de los comercios
metadata_con_tipo = pd.merge (metadata, categorias, left_on= "category", right_on= "Description")
```

```python
# Visualizamos
metadata_con_tipo
```

```python
# Exportamos el archivo csv a ingestar en PowerBi para realizar los análisis
metadata_con_tipo.to_csv("metadata_powerbi.csv")
```
