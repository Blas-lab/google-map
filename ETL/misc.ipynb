```python

def clean_misc(misc):
    if isinstance(misc, str):
        misc_dict = eval(misc)
        for key in misc_dict:
            new_values = [value.replace("-", "_").replace(" ", "_") for value in misc_dict[key]]
            misc_dict[key] = new_values
        return str(misc_dict)
    else:
        return misc

df_misc['misc'] = df_misc['misc'].apply(clean_misc)
```

```python
df_misc
```

```python
import ast

# Define una lista vacía para guardar los valores únicos
valores_unicos = []

# Recorre cada fila del DataFrame
for misc in df_misc['misc']:
    
    # Si es una cadena, conviértela a diccionario
    if isinstance(misc, str):
        misc_dict = ast.literal_eval(misc)
    elif isinstance(misc, dict):
        misc_dict = misc
    else:
        continue
    
    # Recorre cada valor del diccionario y agrega los valores únicos a la lista
    for value_list in misc_dict.values():
        valores_unicos.extend(value_list)
        
# Convierte la lista de valores únicos a un conjunto para eliminar duplicados
valores_unicos = set(valores_unicos)
valores_unicos
```

```python
# Lista de valores a buscar
#valores = ['Good for kids', 'Wheelchair accessible entrance']
# Crear una lista de 1s y 0s para cada valor buscado
encodings = []
for valor in valores_unicos:
    encoding = df_misc['misc'].apply(lambda x: valor in str(x)).astype(int)
    encodings.append(encoding)
    
# Unir los encodings en un DataFrame
df_encodings = pd.concat(encodings, axis=1)

# Renombrar las columnas del DataFrame
df_encodings.columns = valores_unicos
```
